```python
import os

def append_string_to_filenames(folder_path, append_str):
    # Get the list of all files in the folder
    files = os.listdir(folder_path)
    
    for filename in files:
        # Check if the file is a JPG file
        if filename.lower().endswith('.jpg'):
            # Split the file name and extension
            name, ext = os.path.splitext(filename)
            # Create the new file name by appending the string
            new_name = f"{name}{append_str}{ext}"
            # Create the full path for the old and new file names
            old_file = os.path.join(folder_path, filename)
            new_file = os.path.join(folder_path, new_name)
            # Rename the file
            os.rename(old_file, new_file)
            print(f"Renamed: {old_file} -> {new_file}")


append_string_to_filenames('bacterial', '_bacterial')
append_string_to_filenames('fungal', '_fungal')
```

```python
import os

def remove_string_from_filenames(folder_path, remove_str):
    # Get the list of all files in the folder
    files = os.listdir(folder_path)
    
    for filename in files:
        # Check if the file is a JPG file
        if filename.lower().endswith('.jpg'):
            # Split the file name and extension
            name, ext = os.path.splitext(filename)
            # Check if the file name ends with the specified string
            if name.endswith(remove_str):
                # Create the new file name by removing the specified string
                new_name = name[:-len(remove_str)] + ext
                # Create the full path for the old and new file names
                old_file = os.path.join(folder_path, filename)
                new_file = os.path.join(folder_path, new_name)
                # Rename the file
                os.rename(old_file, new_file)
                print(f"Renamed: {old_file} -> {new_file}")


remove_string_from_filenames('bacterial', 'bacterial_bacterial')
remove_string_from_filenames('fungal', 'fungal_fungal')
```
